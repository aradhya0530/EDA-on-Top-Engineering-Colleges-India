```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv('update_data_of_300.csv')
```

```python
df
```

```python
def preprocess_inputs(df):
    df = df.copy()
    df = df.drop('Unnamed: 0', axis = 1)
    
    
    return df
```

```python
X = preprocess_inputs(df)
X
```

```python
X.owner_ship.unique()
# X.loc[298:]['name']
```

```python
plt.figure(figsize = (6,4))
ax = X.owner_ship.value_counts().plot(kind = 'bar', title = 'Public & Private Distribution')
plt.xticks(rotation = 44)
plt.show()
```

```python
## using sns
sns.countplot(x = X['owner_ship'], hue= X['grade'])
plt.show()
```
